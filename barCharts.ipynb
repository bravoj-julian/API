Imported the packages we will need to get the bar graphs to work.
We also imported the csv we created and cleaned with all the data we will need to create our graphs.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df= pd.read_csv('Data_15_22.csv')
df
```

We wanted to make sure our bar graph was divided by both the categories of data and by Education level. So, We created two list to be used to iterate through the values we wanted to search for. This use of list also allowed us to use the code for other categories as well, by simply replacing the values. 
We made this a function to allow for ease to be able to flter the data by year.


```python
# Define the categories and groups
categories = ['Unemployed_total', 'Unemployed_white', 'Unemployed_black', 'Unemployed_asian', 'Unemployed_latino']
groups = ['<HS Diploma', 'HS Diploma', 'Some College', 'Bach+']
y = 2015
def barChart(categories,groups, y):
    # Create a list of positions for the bars
    x = np.arange(len(groups))

    # Define the width of the bars
    width = 0.2  # You can adjust this to change the width of the bars

    # Iterate through the categories and create grouped bar plots
    for i, category in enumerate(categories):
        # Select the relevant row and columns
        data = df[(df['Employment status,\n sex, race, and Hispanic\n or Latino ethnicity'] == category) & (df['Year'] == y)][groups]

        # Extract the values to plot
        data_values = data.values.flatten()

        # Adjust the x position for each group
        x_pos = x + i * width

        # Create the bar plot
        plt.bar(x_pos, data_values, width=width, label=category)

    # Add labels and title
    plt.xlabel('Education Level')
    plt.ylabel('Count')
    plt.title(f'Unemployment Data by Education Level and Race for the year {y}')

    # Set the x-axis ticks and labels
    plt.xticks(x + width, groups)

    # Add legend
    plt.legend()

    # Show the plot
    plt.show() 
```

The following bar charts represent the number of people(*1000) who are unemployed by their race and ethnicity for each Education Level. 

```python
barChart(categories,groups,y)
```
